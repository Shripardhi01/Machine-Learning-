```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
import pandas as pd
```

```python
dataset = fetch_olivetti_faces(shuffle=True)
```

```python
print (dataset.keys())
print (dataset.data.shape)
```

```python
%matplotlib inline

fig = plt.figure()
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
```

```python
for i in range(64): 
    ax=fig.add_subplot(8, 8, i+1, xticks=[], yticks=[]) 
    ax=plt.imshow(dataset.images[i]) 
```

```python
pca = PCA(64) 
print (pca)
```

```python
X, y = dataset.data, dataset.target 
X_proj = pca.fit_transform(X) 
print (X_proj.shape) 
```

```python
plt.scatter(X_proj[:,0], X_proj[:,1],c=y) 
plt.show()
```

```python
pca.components_ 
```

```python
pca.components_.shape
```

```python
pca.explained_variance_ratio_
```

```python
print (np.cumsum(pca.explained_variance_ratio_))
```

```python
fig = plt.figure(figsize=(8,8)) 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05) 
```

```python
for i in range(10): 
    ax = fig.add_subplot(5, 5, i+1, xticks=[], yticks=[]) 
    ax=plt.imshow(np.reshape(pca.components_[i,:], (64,64)), cmap=plt.cm.bone, interpolation='nearest')
```

```python
X_inv_proj = pca.inverse_transform(X_proj) #reshaping as 400 images of 64x64 dimension 
X_proj_img = np.reshape(X_inv_proj,(400,64,64)) 
```

```python
fig = plt.figure(figsize=(6,6)) 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05) 
```

```python
for i in range(64): 
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[]) 
    ax=plt.imshow(X_proj_img[i], cmap=plt.cm.bone, interpolation='nearest') 
```
